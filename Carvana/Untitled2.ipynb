```python
import cPickle as pickle
import os
from PIL import Image
import numpy as np
```

```python
mask = os.listdir("input/train_masks")
```

```python
len(mask)
```

```python
train = os.listdir("input/train_hq")
```

```python
mask
```

```python
train
```

```python
data = []
mask = []
for file in train:
    img = Image.open("input/train_hq/" + file)
    img = img.resize((1024,1024))
    data.append(np.asarray(img))
    mskfile = file.split('.')[0] + "_mask.jpg"
    msk = Image.open("input/train_masks/" + mskfile)
    msk = msk.resize((1024,1024))
    mask.append(np.asarray(msk))
```

```python
print("finished")
```

```python
len(data)
```

```python
trains = {}
trains["img"] = data
trains["mask"] = mask
trains
```

```python
pickle.dump( trains, open( "train", "wb" ) )
```
