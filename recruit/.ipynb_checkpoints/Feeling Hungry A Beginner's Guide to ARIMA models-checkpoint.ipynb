# ** Time Series Primer**

A time series is a structured dataset that can be annual, monthly, or in this case daily. To begin to understand the models, lets outline some key concepts about time-series. 

## ** Time series has to be stationary **
There are 3 conditions that must be fulfilled, otherwise a time series cannot be modeled 

#### ** 1. Constant Mean: AKA no overall trend **
<img src='https://www.analyticsvidhya.com/wp-content/uploads/2015/02/Mean_nonstationary.png' />
#### ** 2. Constant Variance: the magnitude of waviness should remain constant **
<img src='https://www.analyticsvidhya.com/wp-content/uploads/2015/02/Var_nonstationary.png' />
#### ** 3. Frequency: the waves should be evenly spaced **
<img src='https://www.analyticsvidhya.com/wp-content/uploads/2015/02/Cov_nonstationary.png' />

#### ** Another consideration: Seasonailty **
<img src ='https://searchengineland.com/figz/wp-content/seloads/2012/09/High_Seasonality_Pic1.png' />
Seasonality can be on a weekly, monthly, quarterly, or annual basis. 

- Example:the sales of products around Christmas every year
- Example: the number of public transit riders during different days of the week.

### **Simple example: Airplane passengers**
<img src='http://rstatistics.net/wp-content/uploads/2014/09/Seasonal-AirPassengers.png' style='width:500px' />
Note above, the above time series has both a linear upward trend and a seasonal cycle. Luckily the upward trend and seasonality can be addressed with time series techniques


# ** Forecasting in R: ARIMA Models**

<img src='https://i.stack.imgur.com/ux3i7.png' style='width:500px' />

The main technique being explored is **ARIMA** which stands for **Auto Regressive Integrated Moving Average**.

Wikipedia defines ARIMA as: "stands for Autoregressive Integrated Moving Average models. Univariate (single vector) ARIMA is a forecasting technique that projects the future values of a series based entirely on its own inertia. Its main application is in the area of short term forecasting requiring at least 40 historical data points."

The **auto regressive** part of the model determines the **dependence on previous time points**. The **moving average** part of the model determines the **dependence on errors at different points**

## **Load the data****

```r
df_air <- read_csv(file = 'inputair_visit_data.csv')
df_air_store <- read_csv('input/recruit-restaurant-visitor-forecasting/air_store_info.csv')
```

## **Let's Look at the Overall Visitor Data**

```r
par(mfrow=c(2,1), cex=0.7)
df_air %>% 
  group_by(visit_date) %>% 
  summarize(visitors = sum(visitors)) %>% 
  plot(type='l', main='Overall Visitors')

merged <- df_air %>% 
  filter(visit_date > '2016-07-01') %>% 
  dplyr::left_join(df_air_store, by='air_store_id', how='left')

merged_sum <- merged %>% 
  group_by(visit_date) %>% 
  summarize(visitors = sum(visitors)) 

merged_sum %>% 
  plot(type='l', xlab='Year', main='Cut-off at July 2016')
```

#### **Discussion**
We see from the overall dataset, there's definitely a cyclical aspect to the dataset, but overall a steady level of visitors. Each year, there appears to be a very slight upward trend. This is hard to verify because of a strong increase in visitors in the middle of 2016. This strong discontinuity cannot be modeled, so the data used will be cut off at july 2016 or later

## **Let's Model overall visitors: Our first Arima Model**
`arima(merged_train$visitors, order=c(2,1,2), seasonal= list(order=c(1,1,1), period=7))`
#### Comments:
- **train split**: since this is a time series, it can't be completely at random, and the holdout data should avoid irregularities. In this case there's a strong drop in January that we will be sure to avoid when making a decision on cut off
- **`seasonal`** - addresses how many weeks in the past will be examined
- **period** from zooming in we will assume a weekly cycle, or a `period=7` in our arima model. 
-** `order`** - addresses how many days in the past are considered when predicting new visitor values?
-**`arima(212)x()`** - this means that we will go back 2 days for both days and errors as a predictors
-**`arima()x(111)`** - this means that we will go back 1 week for both days and errors as a predictors

```r
merged_train <- merged_sum %>% filter(visit_date <='2017-02-01')
merged_test <- merged_sum %>% filter(visit_date >'2017-02-01')

#print(paste(nrow(merged_sum),nrow(merged_train),nrow(merged_test)))

m <- arima(merged_train$visitors, order=c(2,1,2), seasonal= list(order=c(1,1,1), period=7))
y_pred <- forecast::forecast(m, h=80)

par(mfrow=c(1,1), cex=0.7)
plot(ts(merged_sum$visitors), main="ARIMA model predictions, cut off at Feb 2017")
lines(y_pred$mean, col='red')
```

## **Adjustments for irregular genres - replotted**

**Not all the genres behave the same way** We see that western food, international cuisine, and teppanyaki dare not like the other time series. Either:
- The training period was too different compared to the testing period (International cuisine)
- There's a large irregularity in the trend that throws off the trend (western food, Okonomiyaki)

**Adjust the training cut off and train / test split date to avoide these irregularities**  These issues are addressed by shortening the training period, and adjusting the test train division to avoid big irregularities. In this case, the training data is smaller, only using the first 2017, ignoring 2016. Also the the test period was also shortened accordingly to match.

```r
par(mfrow=c(3,1),cex=0.6)
for (i in c("Western food","International cuisine", "Okonomiyaki/Monja/Teppanyaki") ){
  genre_specific_sum <- genre_sum %>% filter(air_genre_name==i) %>% filter(visit_date > '2017-01-02')
  split_date <- '2017-03-15'
  genre_train <- genre_specific_sum %>% filter(visit_date <= split_date) 
  genre_test <- genre_specific_sum %>% filter(visit_date > split_date)
  
  m <- arima(genre_train$visitors, order=c(2,1,2), seasonal= list(order=c(2,1,2), period=7))
  y_pred <- forecast::forecast(m, h=80)
  
  
  plot(ts(genre_specific_sum$visitors), main=i, xlab='Year', ylab='visitors', xlim=c(0,100))
  lines(y_pred$mean, col='red')
}
```

## **Adjustments for irregular genres - Cafe/Sweets, replotted**

Cafe/sweets also requires to be shortened as well, but even more so. This is to avoid january and the slight increase between feb and march. 

```r
genre_specific_sum <- genre_sum %>% filter(air_genre_name=="Cafe/Sweets") %>% filter(visit_date > '2017-01-02')
split_date <- '2017-02-20'
genre_train <- genre_specific_sum %>% filter(visit_date <= split_date) 
genre_test <- genre_specific_sum %>% filter(visit_date > split_date)

m <- arima(genre_train$visitors, order=c(2,1,2), seasonal= list(order=c(1,1,1), period=7))
y_pred <- forecast::forecast(m, h=80)

par(mfrow=c(2,1),cex=0.6)
genre_sum %>% filter(air_genre_name=='Cafe/Sweets')  %>% select(visitors) %>%  plot(type='l', main = 'Unadjusted')
plot(ts(genre_specific_sum$visitors), main=i, xlab='Year', ylab='visitors', xlim=c(0,100))
lines(y_pred$mean, col='red')
```

## **Import the sample submission as a template**

```r
sampsub <- read_csv('sample_submission.csv') %>% 
  separate(id, c('prefix','store','Date'),'_') %>%
  mutate(store_id = paste0(prefix,'_',store)) %>% 
  mutate(id = paste0(store_id,'_',Date))

#sampsub %>% select(Date) %>% unique %>% arrange(Date)
read_csv('sample_submission.csv') %>% head(4)
```

## **Make the Genre Predictions**

We will use the same loop as before and store the predictions per genre.

```r
graph_list <- list()
index <- 1
pred_dates <- seq.Date(as.Date('2017-04-23'), as.Date('2017-05-31'),by=1)


for (i in genre_unique){
  print(i)
  if (i %in% c("Western food","International cuisine", "Okonomiyaki/Monja/Teppanyaki")){
    genre_specific_sum <- genre_sum %>% 
      filter(air_genre_name==i) %>% 
      filter(visit_date > '2017-01-02')
    
    split_date <- '2017-03-15'
    
  } else if(i == "Cafe/Sweets"){
    genre_specific_sum <- genre_sum %>% 
      filter(air_genre_name=="Cafe/Sweets") %>% 
      filter(visit_date > '2017-01-02')
    
    split_date <- '2017-02-20'
  } else{
    genre_specific_sum <- genre_sum %>% filter(air_genre_name==i)
    split_date <- '2017-02-01'
  }
  
  genre_train <- genre_specific_sum

  m <- arima(genre_train$visitors, order=c(2,1,2), seasonal= list(order=c(1,1,1), period=7))
  y_pred <- forecast::forecast(m, h=39)
  graph_list[[index]] <- data.frame(air_genre_name=i, visit_date=pred_dates, visitors=y_pred$mean, stringsAsFactors = FALSE)
  index <- index +1
```

All the individual predictions are stored in individual dataframes and we will stitch them all together

```r
base_df <- graph_list[[1]] %>% 
  bind_rows(graph_list[[2]]) %>% 
  bind_rows(graph_list[[3]]) %>% 
  bind_rows(graph_list[[4]]) %>% 
  bind_rows(graph_list[[5]]) %>% 
  bind_rows(graph_list[[6]]) %>% 
  bind_rows(graph_list[[7]]) %>% 
  bind_rows(graph_list[[8]]) %>% 
  bind_rows(graph_list[[9]]) %>% 
  bind_rows(graph_list[[10]]) %>% 
  bind_rows(graph_list[[11]]) %>% 
  bind_rows(graph_list[[12]]) %>% 
  bind_rows(graph_list[[13]]) %>% 
  bind_rows(graph_list[[14]])
```

## **Make the Allocation Table by Store**
This is a 2 step summary process. First the visitor totals are calculated first per genre, then second by store. Then we will create the table outlined above, calculating the stores % of visitors of the genre total.

```r
genre_totals <- genre_sum %>% 
  mutate(yr_m = format(visit_date,"%Y-%m")) %>%
  filter(yr_m %in% c('2017-04','2017-03')) %>%
  group_by(air_genre_name) %>% 
  summarize(visitors=sum(visitors))

store_alloc <- merged %>% 
  mutate(yr_m = format(visit_date,"%Y-%m")) %>%
  filter(yr_m %in% c('2017-04','2017-03')) %>%
  group_by(air_store_id, air_genre_name) %>% 
  summarize(store_visitors =sum(visitors)) %>% 
  left_join(genre_totals, by=c('air_genre_name'='air_genre_name')) %>% 
  mutate(pct=store_visitors / visitors) %>% 
  select(air_store_id, air_genre_name, pct) %>% 
  arrange(air_genre_name)
```

## **Allocate the Genre Forecast down to the Store Level and save as CSV**

We will take advantage of the left-join to create all the extra rows. Left joining the smaller genre totals table x store allocation table, and will make a larger table, which is referred to a cartesian join. Finally the results are saved to CSV

```r

pred_subs <- base_df %>% 
  left_join(store_alloc, how='left', by="air_genre_name") %>% 
  mutate(store_visitors = visitors*pct) %>% 
  mutate(id=paste0(air_store_id,'_',visit_date)) %>%
  select(id, store_visitors)


out_preds <- sampsub %>%  left_join(pred_subs, by="id") %>% 
  select(id, store_visitors)

colnames(out_preds) <- c("id","visitors")

out_preds %>% write_csv("arima_alloc.csv")
```
